```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
randomForest = RandomForestClassifier(n_estimators=100, max_depth=5, min_samples_leaf=50, class_weight='balanced')
```

```python
features = pd.read_csv('CreacionDeFeatures/trainingSet.csv')
testing = pd.read_csv('CreacionDeFeatures/testingSet.csv')
testing = testing.set_index('person')

labels = pd.read_csv('CreacionDeFeatures/Data/labels_training_set.csv')
```

```python
features = features.set_index('person')

features.head()
```

```python
labels = labels.set_index('person')

labels.head()
```

```python
data = features.merge(labels, left_index=True, right_index=True)

data.head()
```

```python
features, labels = data.iloc[:,:-1],data.iloc[:,-1]
```

```python
features.head()
```

```python
labels.head()

#Estan en el mismo orden
```

```python
features_train, features_test, labels_train, labels_test = \
    train_test_split(features, labels, test_size=0.3, random_state=153)
```

```python
randomForest.fit(features_train, labels_train)
```

```python
randomForest.score(features_test,labels_test)
```

```python
preds = randomForest.predict(testing)
```

```python
randomForestFinal = RandomForestClassifier(n_estimators=100, max_depth=5, min_samples_leaf=50, class_weight='balanced')
```

```python
randomForestFinal.fit(features, labels)
```

```python
preds = randomForestFinal.predict(testing)
```
