```python
import pandas as pd
import numpy as np
```

```python
logsTrocafone = pd.read_csv('Data/events_up_to_01062018.csv', low_memory = False)

trainingSet = pd.read_csv('trainingSet.csv')
trainingSet = trainingSet.set_index('person')

testingSet = pd.read_csv('testingSet.csv')
testingSet = testingSet.set_index('person')

trainingSet.head()
```

```python
logsTrocafoneCompras = logsTrocafone.loc[logsTrocafone['event'] == 'conversion', :]

logsTrocafoneCompras.head()
```

```python
cantidadDeCompras = logsTrocafoneCompras.groupby('person').agg({'event':'count'})

cantidadDeCompras.head()
```

```python
nuevoTrainingSet = trainingSet.merge(cantidadDeCompras, how='left', left_index=True, right_index=True)

nuevoTrainingSet.info()
```

```python
nuevoTestingSet = testingSet.merge(cantidadDeCompras, how='left', left_index=True, right_index=True)

nuevoTestingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.fillna(0)

nuevoTrainingSet.info()
```

```python
nuevoTestingSet = nuevoTestingSet.fillna(0)

nuevoTestingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.rename(columns={"event": "CantidadCompras"})

nuevoTestingSet = nuevoTestingSet.rename(columns={"event": "CantidadCompras"})
```
