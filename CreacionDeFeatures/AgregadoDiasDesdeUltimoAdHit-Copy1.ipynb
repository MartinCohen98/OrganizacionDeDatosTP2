```python
import pandas as pd
import numpy as np
import datetime
```

```python
logsTrocafone = pd.read_csv('Data/events_up_to_01062018.csv', low_memory = False)

trainingSet = pd.read_csv('trainingSet.csv')
trainingSet = trainingSet.set_index('person')

testingSet = pd.read_csv('testingSet.csv')
testingSet = testingSet.set_index('person')

trainingSet.head()
```

```python
logsTrocafone['timestamp'] = pd.to_datetime(logsTrocafone['timestamp'], errors = 'coerce')
logsTrocafone.head()
```

```python
logsTrocafone['timestamp'] = logsTrocafone['timestamp'].dt.date

logsTrocafone.head()
```

```python
searchHits = logsTrocafone.loc[logsTrocafone['event'] == 'search engine hit', :]

searchHits.head()
```

```python
ultimoEventoPorPersona = searchHits.groupby('person').agg({'timestamp':'max'})

ultimoEventoPorPersona.head()
```

```python
ultimoEventoPorPersona['timestamp'] =  datetime.date(2018, 6, 1) - ultimoEventoPorPersona['timestamp']

ultimoEventoPorPersona['timestamp'].head()
```

```python
ultimoEventoPorPersona['timestamp'] = ultimoEventoPorPersona['timestamp'] / np.timedelta64(1, 'D')

ultimoEventoPorPersona.head()
```

```python
nuevoTrainingSet = trainingSet.merge(ultimoEventoPorPersona, how='left', left_index=True, right_index=True)

nuevoTrainingSet.info()
```

```python
nuevoTestingSet = testingSet.merge(ultimoEventoPorPersona, how='left', left_index=True, right_index=True)

nuevoTestingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.fillna(1000)

#Agrego un numero muy grande significando que na hacen un checkout hace mucho

nuevoTrainingSet.info()
```

```python
nuevoTestingSet = nuevoTestingSet.fillna(1000)

#Agrego un numero muy grande significando que na hacen un checkout hace mucho

nuevoTestingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.rename(columns={"timestamp": "DiasDesdeUltimoSearchHit"})

nuevoTrainingSet.head()
```

```python
nuevoTestingSet = nuevoTestingSet.rename(columns={"timestamp": "DiasDesdeUltimoSearchHit"})

nuevoTestingSet.head()
```
