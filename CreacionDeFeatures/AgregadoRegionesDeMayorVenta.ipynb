```python
df = logsTrocafone[['person', 'region']].dropna()
df = df.loc[df['region'] != 'Unknown', :]
df.head(10)
```

Veo desde que region la persona visita mas

```python
df = df.groupby(['person']).agg({'region' : lambda x:x.value_counts().index[0]})
df = df.reset_index()
```

```python
logsTrocafone = logsTrocafone.merge(df, how='left', left_index=True, right_index=True)
```

```python
logsTrocafone.head()
```

```python
trainingSet = pd.read_csv('trainingSet2.csv')
trainingSet = trainingSet.set_index('person')
```

```python
trainingSet.shape
```

```python
trainingSet = trainingSet.merge(logsTrocafone, how='left', left_index=True, right_index=True)
trainingSet = trainingSet.fillna(0)
trainingSet.head()
```
