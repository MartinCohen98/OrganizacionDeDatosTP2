#### Vemos si los registros para cada sku (modelo) son un buen feature, ya que algun producto posdria estar de moda.

```python
logsTrocafone['sku_count'] = 1
```

```python
dftmp = logsTrocafone.loc[(logsTrocafone['sku'] == logsTrocafone['sku']), \
                          ['person', 'sku', 'sku_count']]
dftmp = dftmp.set_index('person')
dftmp.head(20)
```

```python
trainingSet = pd.read_csv('trainingSet2.csv')
trainingSet = trainingSet.set_index('person')
```

```python
trs = pd.DataFrame(index=trainingSet.index)
trs.head()
```

```python
dftmp_train = trs.merge(dftmp, how='left', left_index=True, right_index=True)
dftmp_train = dftmp_train.fillna(0)
dftmp_train.head()
```

```python
dftmp_train = dftmp_train.pivot_table(index='person',columns='sku',aggfunc=sum)
dftmp_train.fillna(0,inplace=True)
dftmp_train.columns = dftmp_train.columns.droplevel()
dftmp_train.columns.name = None
dftmp_train.reset_index(inplace=True)
print(len(dftmp_train.columns))
dftmp_train = dftmp_train.set_index('person')
dftmp_train.head()
```

```python
dftmp_train.info()
```

```python
for column in dftmp_train.columns[45:]:
    print(dftmp_train[column].sum())
```

```python
testingSet = pd.read_csv('testingSet2.csv')
testingSet = testingSet.set_index('person')
```

```python
tss = pd.DataFrame(index=testingSet.index)
tss.head()
```

```python
dftmp_test = tss.merge(dftmp, how='left', left_index=True, right_index=True)
dftmp_test = dftmp_test.fillna(0)
dftmp_test.head()
```

```python
dftmp_test = dftmp_test.pivot_table(index='person',columns='sku',aggfunc=sum)
dftmp_test.fillna(0,inplace=True)
dftmp_test.columns = dftmp_test.columns.droplevel()
dftmp_test.columns.name = None
dftmp_test.reset_index(inplace=True)
print(len(dftmp_test.columns))
dftmp_test = dftmp_test.set_index('person')
dftmp_test.head()
```

```python
dftmp_test.info()
```

```python
for column in dftmp_test.columns[45:]:
    print(dftmp_test[column].sum())
```

```python
nuevoTrainingSet = trainingSet.merge(dftmp_train, how='left', left_index=True, right_index=True)
nuevoTrainingSet = nuevoTrainingSet.fillna(0)
nuevoTrainingSet.head()
```

```python
nuevoTrainingSet.info()
```

```python
nuevoTestingSet = testingSet.merge(dftmp_test, how='left', left_index=True, right_index=True)
nuevoTestingSet = nuevoTestingSet.fillna(0)
nuevoTestingSet.head()
```

```python
nuevoTestingSet.info()
```

#### Bucamos ver si quienes compraron de determinada 'condition' vuelven a comprar

```python
logsTrocafone['timestamp'] = pd.to_datetime(logsTrocafone['timestamp'], errors = 'coerce')
logsTrocafone['timestamp'] = logsTrocafone['timestamp'].dt.date
```

```python
logsTrocafone['Hace'] = logsTrocafone['timestamp'] >= datetime.date(2018,5,1)
```

```python
logsTrocafone['ventas'] = 1
df = logsTrocafone
df = df.loc[(df['event'] == 'conversion') & (df['Hace'] == True), :]
df.head()
```

```python
df.info()
```

```python
dftmp = df.loc[:, ['person', 'condition', 'ventas']]
dftmp = dftmp.groupby(['person','condition']).agg({'ventas': 'sum'})
dftmp.head(10)
```

```python
dftmp = dftmp.unstack(level=-1)
dftmp.fillna(0,inplace=True)
dftmp.columns = dftmp.columns.droplevel()
dftmp.columns.name = None
dftmp.reset_index(inplace=True)
dftmp = dftmp.set_index('person')
dftmp.head(10)
```

```python
nuevoTrainingSet = trainingSet.merge(dftmp, how='left', left_index=True, right_index=True)
nuevoTrainingSet = nuevoTrainingSet.fillna(0)
nuevoTrainingSet['Bom'] = nuevoTrainingSet['Bom'].astype(int)
nuevoTrainingSet['Bom - Sem Touch ID'] = nuevoTrainingSet['Bom - Sem Touch ID'].astype(int)
nuevoTrainingSet['Excelente'] = nuevoTrainingSet['Excelente'].astype(int)
nuevoTrainingSet['Muito Bom'] = nuevoTrainingSet['Muito Bom'].astype(int)
nuevoTrainingSet['Novo'] = nuevoTrainingSet['Novo'].astype(int)
nuevoTrainingSet.head()
```

```python
nuevoTestingSet = testingSet.merge(dftmp, how='left', left_index=True, right_index=True)
nuevoTestingSet = nuevoTestingSet.fillna(0)
nuevoTestingSet['Bom'] = nuevoTestingSet['Bom'].astype(int)
nuevoTestingSet['Bom - Sem Touch ID'] = nuevoTestingSet['Bom - Sem Touch ID'].astype(int)
nuevoTestingSet['Excelente'] = nuevoTestingSet['Excelente'].astype(int)
nuevoTestingSet['Muito Bom'] = nuevoTestingSet['Muito Bom'].astype(int)
nuevoTestingSet['Novo'] = nuevoTestingSet['Novo'].astype(int)
nuevoTestingSet.head()
```

```python
nuevoTrainingSet.to_csv('trainingSet3.csv')
#Este comando reemplaza el training set existente con la feature agregada
nuevoTestingSet.to_csv('testingSet3.csv')
#Este comando reemplaza el testing set existente con la feature agregada
```
