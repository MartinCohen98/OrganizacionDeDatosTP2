```python
import pandas as pd
import numpy as np
import datetime
```

```python
logsTrocafone = pd.read_csv('Data/events_up_to_01062018.csv', low_memory = False)

trainingSet = pd.read_csv('trainingSet.csv')
trainingSet = trainingSet.set_index('person')

trainingSet.head()
```

```python
logsTrocafone['timestamp'] = pd.to_datetime(logsTrocafone['timestamp'], errors = 'coerce')
logsTrocafone.head()
```

```python
logsTrocafone['timestamp'] = logsTrocafone['timestamp'].dt.date

logsTrocafone.head()
```

```python
logsTrocafone['Mes1'] = ((logsTrocafone['timestamp'] >= datetime.date(2018,1,1)) & (logsTrocafone['timestamp'] <= datetime.date(2018,1,31)))

logsTrocafone['Mes2'] = ((logsTrocafone['timestamp'] >= datetime.date(2018,2,1)) & (logsTrocafone['timestamp'] <= datetime.date(2018,2,28)))

logsTrocafone['Mes3'] = ((logsTrocafone['timestamp'] >= datetime.date(2018,3,1)) & (logsTrocafone['timestamp'] <= datetime.date(2018,3,31)))

logsTrocafone['Mes4'] = ((logsTrocafone['timestamp'] >= datetime.date(2018,4,1)) & (logsTrocafone['timestamp'] <= datetime.date(2018,4,30)))

logsTrocafone['Mes5'] = ((logsTrocafone['timestamp'] >= datetime.date(2018,5,1)) & (logsTrocafone['timestamp'] <= datetime.date(2018,5,31)))

logsTrocafone.head()
```

```python
ventas = logsTrocafone.loc[logsTrocafone['event'] == 'conversion', :]

ventas.head()
```

```python
ventasMes1 = ventas.loc[ventas['Mes1'] == True, :]

ventasMes2 = ventas.loc[ventas['Mes2'] == True, :]

ventasMes3 = ventas.loc[ventas['Mes3'] == True, :]

ventasMes4 = ventas.loc[ventas['Mes4'] == True, :]

ventasMes5 = ventas.loc[ventas['Mes5'] == True, :]
```

```python
ventasMes1PorPersona = ventasMes1.groupby('person').agg({'event':'count'})

ventasMes1PorPersona = ventasMes1PorPersona.rename(columns={"event": "VentasMes1"})

ventasMes2PorPersona = ventasMes2.groupby('person').agg({'event':'count'})

ventasMes2PorPersona = ventasMes2PorPersona.rename(columns={"event": "VentasMes2"})

ventasMes3PorPersona = ventasMes3.groupby('person').agg({'event':'count'})

ventasMes3PorPersona = ventasMes3PorPersona.rename(columns={"event": "VentasMes3"})

ventasMes4PorPersona = ventasMes4.groupby('person').agg({'event':'count'})

ventasMes4PorPersona = ventasMes4PorPersona.rename(columns={"event": "VentasMes4"})

ventasMes5PorPersona = ventasMes5.groupby('person').agg({'event':'count'})

ventasMes5PorPersona = ventasMes5PorPersona.rename(columns={"event": "VentasMes5"})
```

```python
ventasPorMes = ventasMes1PorPersona.merge(ventasMes2PorPersona, how='outer', left_index=True, right_index=True)

ventasPorMes = ventasPorMes.merge(ventasMes3PorPersona, how='outer', left_index=True, right_index=True)

ventasPorMes = ventasPorMes.merge(ventasMes4PorPersona, how='outer', left_index=True, right_index=True)

ventasPorMes = ventasPorMes.merge(ventasMes5PorPersona, how='outer', left_index=True, right_index=True)

ventasPorMes.head()
```

```python
ventasPorMes = ventasPorMes.fillna(0)

ventasPorMes.head()
```

```python
ventasPorMes['PromedioVentasPorMes'] = (ventasPorMes['VentasMes1'] + ventasPorMes['VentasMes2'] + ventasPorMes['VentasMes3']\
                                  + ventasPorMes['VentasMes4'] + ventasPorMes['VentasMes5'])/5

ventasPorMes.head()
```

```python
ventasPorMes = ventasPorMes.drop('VentasMes5', axis=1)

ventasPorMes.head()
```

```python
nuevoTrainingSet = trainingSet.merge(ventasPorMes, how='left', left_index=True, right_index=True)

nuevoTrainingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.fillna(0)

nuevoTrainingSet.info()
```
