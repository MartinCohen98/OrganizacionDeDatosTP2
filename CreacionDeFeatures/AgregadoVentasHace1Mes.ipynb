```python
import pandas as pd
import numpy as np
import datetime
```

```python
logsTrocafone = pd.read_csv('Data/events_up_to_01062018.csv', low_memory = False)

trainingSet = pd.read_csv('trainingSet.csv')
trainingSet = trainingSet.set_index('person')

trainingSet.head()
```

```python
logsTrocafone['timestamp'] = pd.to_datetime(logsTrocafone['timestamp'], errors = 'coerce')
logsTrocafone.head()
```

```python
logsTrocafone['timestamp'].max()
```

```python
logsTrocafone['timestamp'] = logsTrocafone['timestamp'].dt.date

logsTrocafone.head()
```

```python
logsTrocafone['Hace1Mes'] = logsTrocafone['timestamp'] >= datetime.date(2018,5,1)
logsTrocafone.head()
```

```python
logsHaceUnMes = logsTrocafone.loc[logsTrocafone['Hace1Mes'] == True, :]

logsHaceUnMes.head()
```

```python
ventasHaceUnMes = logsHaceUnMes.loc[logsHaceUnMes['event'] == 'conversion', :]

ventasHaceUnMes.head()
```

```python
ventasHaceUnMesPorPersona = ventasHaceUnMes.groupby('person').agg({'event':'count'})

ventasHaceUnMesPorPersona.head()
```

```python
nuevoTrainingSet = trainingSet.merge(ventasHaceUnMesPorPersona, how='left', left_index=True, right_index=True)

nuevoTrainingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.fillna(0)

nuevoTrainingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.rename(columns={"event": "VentasHaceUnMes"})

nuevoTrainingSet.head()
```
