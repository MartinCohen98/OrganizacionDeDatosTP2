```python
import pandas as pd
import numpy as np
import datetime
```

```python
logsTrocafone = pd.read_csv('Data/events_up_to_01062018.csv', low_memory = False)

trainingSet = pd.read_csv('trainingSet.csv')
trainingSet = trainingSet.set_index('person')

testingSet = pd.read_csv('testingSet.csv')
testingSet = testingSet.set_index('person')

trainingSet.head()
```

```python
returning = logsTrocafone.loc[logsTrocafone['new_vs_returning'] == 'Returning', :]

returning.head()
```

```python
returning['ColumnaDe1'] = 1

returning.head()
```

```python
personasQueVolvieron = returning.groupby('person').agg({'ColumnaDe1':'max'})

personasQueVolvieron.head()
```

```python
nuevoTrainingSet = trainingSet.merge(personasQueVolvieron, how='left', left_index=True, right_index=True)

nuevoTrainingSet.info()
```

```python
nuevoTestingSet = testingSet.merge(personasQueVolvieron, how='left', left_index=True, right_index=True)

nuevoTestingSet.info()
```

```python
nuevoTestingSet = nuevoTestingSet.fillna(0)

nuevoTestingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.fillna(0)

nuevoTrainingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.rename(columns={"ColumnaDe1": "EntraronMasDeUnaVez"})

nuevoTestingSet = nuevoTestingSet.rename(columns={"ColumnaDe1": "EntraronMasDeUnaVez"})
```
