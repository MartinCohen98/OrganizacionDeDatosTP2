```python
import pandas as pd
import numpy as np
```

```python
logsTrocafone = pd.read_csv('Data/events_up_to_01062018.csv', low_memory = False)

trainingSet = pd.read_csv('trainingSet.csv')
trainingSet = trainingSet.set_index('person')

trainingSet.head()
```

```python
logsTrocafoneCheckouts = logsTrocafone.loc[logsTrocafone['event'] == 'checkout', :]

logsTrocafoneCheckouts.head()
```

```python
cantidadDeCheckouts = logsTrocafoneCheckouts.groupby('person').agg({'event':'count'})

cantidadDeCheckouts.head()
```

```python
nuevoTrainingSet = trainingSet.merge(cantidadDeCheckouts, how='left', left_index=True, right_index=True)

nuevoTrainingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.fillna(0)

nuevoTrainingSet.info()
```

```python
nuevoTrainingSet.head()
```

```python
nuevoTrainingSet = nuevoTrainingSet.rename(columns={"event": "CantidadCheckouts"})
```
