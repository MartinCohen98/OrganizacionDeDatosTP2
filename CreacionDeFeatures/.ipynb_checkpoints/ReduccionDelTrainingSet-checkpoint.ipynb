```python
import pandas as pd
import numpy as np
```

```python
features = pd.read_csv('trainingSet.csv')
features = features.set_index('person')

labels = pd.read_csv('Data/labels_training_set.csv')
labels = labels.set_index('person')
```

```python
labels.count()
```

```python
labels.sum()
```

```python
#Hay muchos mas 0 que 1, hago un sample de los 0 para igualar la cantidad

ceros = labels.loc[labels['label'] == 0, :]

ceros.sum()
```

```python
ceros.count()
```

```python
trainingSetReducido = ceros.sample(n = 980, replace = False, random_state = 143)

trainingSetReducido.head(40)
```

```python
trainingSetReducido.count()
```

```python
trainingSetReducido.sum()
```

```python
cerosReducido = trainingSetReducido
```

```python
unos = labels.loc[labels['label'] == 1, :]
```

```python
unos.sum()
```

```python
unos.count()
```

```python
labelsReducidas = pd.concat([unos, cerosReducido])
```

```python
labelsReducidas.head()
```

```python
labelsReducidas.sum()
```

```python
labelsReducidas.count()
```

```python
data = features.merge(labelsReducidas, left_index=True, right_index=True)

data
```

```python
featuresReducidas, labelsReducidas = data.iloc[:,:-1],data.iloc[:,-1]
```

```python
featuresReducidas
```

```python
labelsReducidas
```

```python
featuresReducidas.to_csv('trainingSetReducido.csv')
```

```python
labelsReducidas.to_csv('labelsTrainingReducidas.csv')
```
