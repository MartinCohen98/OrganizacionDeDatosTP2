```python
import pandas as pd
import numpy as np
```

```python
logsTrocafone = pd.read_csv('Data/events_up_to_01062018.csv', low_memory = False)

trainingSet = pd.read_csv('trainingSet.csv')
trainingSet = trainingSet.set_index('person')

testingSet = pd.read_csv('testingSet.csv')
testingSet = testingSet.set_index('person')

trainingSet.head()
```

```python
logsTrocafone['device_type'].value_counts()

logsTrocafone['ColumnaDe1'] = 1
```

```python
logsSmartphone = logsTrocafone.loc[logsTrocafone['device_type'] == 'Smartphone', :]

logsComputadora = logsTrocafone.loc[logsTrocafone['device_type'] == 'Computer', :]

logsTablet = logsTrocafone.loc[logsTrocafone['device_type'] == 'Tablet', :]
```

```python
logsSmartphone.head()
```

```python
logsComputadora.head()
```

```python
logsTablet.head()
```

```python
entroConSmartphone = logsSmartphone.groupby('person').agg({'ColumnaDe1':'max'})

entroConSmartphone = entroConSmartphone.rename(columns={"ColumnaDe1": "EntraronConSmartphone"})

entroConComputadora = logsComputadora.groupby('person').agg({'ColumnaDe1':'max'})

entroConComputadora = entroConComputadora.rename(columns={"ColumnaDe1": "EntraronConComputadora"})

entroConTablet = logsTablet.groupby('person').agg({'ColumnaDe1':'max'})

entroConTablet = entroConTablet.rename(columns={"ColumnaDe1": "EntraronConTablet"})
```

```python
entradaPorDispositivo = entroConComputadora.merge(entroConSmartphone, how='outer', left_index=True, right_index=True)

entradaPorDispositivo = entradaPorDispositivo.merge(entroConTablet, how='outer', left_index=True, right_index=True)

entradaPorDispositivo.head()
```

```python
nuevoTestingSet = testingSet.merge(entradaPorDispositivo, how='left', left_index=True, right_index=True)

nuevoTestingSet.info()
```

```python
nuevoTestingSet = nuevoTestingSet.fillna(0)

nuevoTestingSet.info()
```

```python
nuevoTrainingSet = trainingSet.merge(entradaPorDispositivo, how='left', left_index=True, right_index=True)

nuevoTrainingSet.info()
```

```python
nuevoTrainingSet = nuevoTrainingSet.fillna(0)

nuevoTrainingSet.info()
```
