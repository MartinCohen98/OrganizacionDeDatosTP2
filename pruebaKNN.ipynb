```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
features = pd.read_csv('CreacionDeFeatures/trainingSetReducido.csv')

testing = pd.read_csv('CreacionDeFeatures/testingSet.csv')
testing = testing.set_index('person')

labels = pd.read_csv('CreacionDeFeatures/labelsTrainingReducidas.csv')
```

```python
features = features.set_index('person')

features.head()
```

```python
labels = labels.set_index('person')

labels.head()
```

```python
data = features.merge(labels, left_index=True, right_index=True)

data.head()
```

```python
features, labels = data.iloc[:,:-1],data.iloc[:,-1]
```

```python
features.head()
```

```python
labels.head()

#Estan en el mismo orden
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(features, labels)
```

```python
preds = knn.predict(testing)
```
